```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Twitter.csv')
data.head()
```

```python
data.shape
```

```python
data['2401'].value_counts()
```

```python
data['Borderlands'].value_counts()
```

```python
plt.pie(x=data['Positive'].value_counts().values, labels=data['Positive'].value_counts().index)
plt.title('Positive content vs Negative content vs Neutral content vs Irrelevant content')
plt.show()
```

```python
sns.barplot(y=data['2401'].value_counts().index, x=data['2401'].value_counts().values)
plt.xlabel('count')
plt.ylabel('Total')
plt.title('Count vs Total')
plt.show()
```

```python
sns.barplot(x=data['Borderlands'].value_counts().values, y=data['Borderlands'].value_counts().index)
plt.xlabel('Count')
plt.ylabel('Social Medias')
plt.title('Count vs Social Medias')
plt.show()
```

```python
sns.catplot(x='Borderlands',hue='Positive',kind='count', height=6,aspect=1,data=data, palette = ["#F4EC5A","#A5C9CA", "#C3AD21", "#F10898"])
```
